```python
# All values should be obtained by using (complex - receptor - ligand)
######### These values should be generalized to work with any .xvg output - no editing required! ##########
comp0 = 0
compf = int(len(time1)/3)
rec0 = compf 
recf = int(2*len(time1)/3)
lig0 = recf
###########################################################################################################

internal1 = bond1 + angle1 + dihed1
Elec1 = eel1 + egb1 + eel_1_4_1
VdW1 = vdwaals1 + vdw_1_4_1 + esurf1

internal2 = bond2 + angle2 + dihed2
Elec2 = eel2 + egb2 + eel_1_4_2
VdW2 = vdwaals2 + vdw_1_4_2 + esurf2

internal3 = bond3 + angle3 + dihed3
Elec3 = eel3 + egb3 + eel_1_4_3
VdW3 = vdwaals3 + vdw_1_4_3 + esurf3

internal4 = bond4 + angle4 + dihed4
Elec4 = eel4 + egb4 + eel_1_4_4
VdW4 = vdwaals4 + vdw_1_4_4 + esurf4

internal5 = bond5 + angle5 + dihed5
Elec5 = eel5 + egb5 + eel_1_4_5
VdW5 = vdwaals5 + vdw_1_4_5 + esurf5

internal_final1 = internal1[comp0:compf] - internal1[rec0:recf] - internal1[lig0::]
Elec_final1 = Elec1[comp0:compf] - Elec1[rec0:recf] - Elec1[lig0::]
VdW_final1 = VdW1[comp0:compf] - VdW1[rec0:recf] - VdW1[lig0::]

internal_final2 = internal2[comp0:compf] - internal2[rec0:recf] - internal2[lig0::]
Elec_final2 = Elec2[comp0:compf] - Elec2[rec0:recf] - Elec2[lig0::]
VdW_final2 = VdW2[comp0:compf] - VdW2[rec0:recf] - VdW2[lig0::]

internal_final3 = internal3[comp0:compf] - internal3[rec0:recf] - internal3[lig0::]
Elec_final3 = Elec3[comp0:compf] - Elec3[rec0:recf] - Elec3[lig0::]
VdW_final3 = VdW3[comp0:compf] - VdW3[rec0:recf] - VdW3[lig0::]

internal_final4 = internal4[comp0:compf] - internal4[rec0:recf] - internal4[lig0::]
Elec_final4 = Elec4[comp0:compf] - Elec4[rec0:recf] - Elec4[lig0::]
VdW_final4 = VdW4[comp0:compf] - VdW4[rec0:recf] - VdW4[lig0::]

internal_final5 = internal5[comp0:compf] - internal5[rec0:recf] - internal5[lig0::]
Elec_final5 = Elec5[comp0:compf] - Elec5[rec0:recf] - Elec5[lig0::]
VdW_final5 = VdW5[comp0:compf] - VdW5[rec0:recf] - VdW5[lig0::]

internal_final = np.concatenate([internal_final1,internal_final2,internal_final3,internal_final4,internal_final5])
Elec_final = np.concatenate([Elec_final1,Elec_final2,Elec_final3,Elec_final4,Elec_final5])
VdW_final = np.concatenate([VdW_final1,VdW_final2,VdW_final3,VdW_final4,VdW_final5])

# Save results to a space-separated file
output_file1 = "EnoB1_pymbar_data.txt"
with open(output_file1, "w") as f:
    f.write("# Internal Elec VdW\n")  # Header
    for i in range(len(internal_final1)):
        f.write(f"{i} {internal_final1[i]:.6f} {Elec_final1[i]:.6f} {VdW_final1[i]:.6f}\n")
print(f"Data written to {output_file1}")

output_file2 = "EnoB2_pymbar_data.txt"
with open(output_file2, "w") as f:
    f.write("# Internal Elec VdW\n")  # Header
    for i in range(len(internal_final2)):
        f.write(f"{i} {internal_final2[i]:.6f} {Elec_final2[i]:.6f} {VdW_final2[i]:.6f}\n")
print(f"Data written to {output_file2}")

output_file3 = "EnoB3_pymbar_data.txt"
with open(output_file3, "w") as f:
    f.write("# Internal Elec VdW\n")  # Header
    for i in range(len(internal_final3)):
        f.write(f"{i} {internal_final3[i]:.6f} {Elec_final3[i]:.6f} {VdW_final3[i]:.6f}\n")
print(f"Data written to {output_file3}")

output_file4 = "EnoB4_pymbar_data.txt"
with open(output_file4, "w") as f:
    f.write("# Internal Elec VdW\n")  # Header
    for i in range(len(internal_final4)):
        f.write(f"{i} {internal_final4[i]:.6f} {Elec_final4[i]:.6f} {VdW_final4[i]:.6f}\n")
print(f"Data written to {output_file4}")

output_file5 = "EnoB5_pymbar_data.txt"
with open(output_file5, "w") as f:
    f.write("# Internal Elec VdW\n")  # Header
    for i in range(len(internal_final5)):
        f.write(f"{i} {internal_final5[i]:.6f} {Elec_final5[i]:.6f} {VdW_final5[i]:.6f}\n")
print(f"Data written to {output_file5}")
```

```python
def Error_prop(x,y,z):
    a = stat.stdev(x)
    b = stat.stdev(y)
    c = stat.stdev(z)
    error = (a**2 + b**2 + c**2)**0.5
    return float(error)
```

```python
print('Your final Internal Energy is '+str(stat.mean(internal_final))+' +/- '+str(stat.stdev(internal_final)/2.2)+'.')
print('Your final Electrostatic Energy is '+str(stat.mean(Elec_final))+' +/- '+str(stat.stdev(Elec_final)/2.2)+'.')
print('Your final Van der Waals Energy is '+str(stat.mean(VdW_final))+' +/- '+str(stat.stdev(VdW_final)/2.2)+'.')
print('Your total energy is '+str(stat.mean(internal_final+VdW_final+Elec_final))+' +/- '+str(float(Error_prop(VdW_final,internal_final,Elec_final))/2.2)+'.')
```
