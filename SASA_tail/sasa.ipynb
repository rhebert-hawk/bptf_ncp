```python
font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'normal',
        'size': 12,
        }

x = np.loadtxt("Nucmsurfall.dat", usecols=(1), unpack=True, comments=('#','@','&'))
z = np.loadtxt("NucEsurfall.dat", usecols=(1), unpack=True, comments=('#','@','&'))
b = np.loadtxt("EnoBsurfall.dat", usecols=(1), unpack=True, comments=('#','@','&'))

plt.hist(x, alpha=0.5, bins=300, density=True, linewidth=3, label='NCP', histtype='step')

plt.hist(b, alpha=0.5, bins=300, density=True, linewidth=3, label='PHD + NCP', histtype='step')

plt.hist(z, alpha=0.5, bins=300, density=True, linewidth=3, label='BPTF + NCP', histtype='step')


plt.xlabel('Solvent Accessible Surface Area ($\AA^2$)', fontdict=font)
plt.ylabel('Frequency', fontdict=font)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('H3K4me3 Solvent Accessible Surface Area', fontdict=font, fontweight='bold')
plt.legend(prop={'size': 12}, loc='upper right')
plt.tight_layout()
plt.savefig('Tail_SASA.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.gridspec as gridspec
from PIL import Image
font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'bold',
        'size': 10,
        }

#x,y = np.loadtxt("NTeqrg.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))

#c,d = np.loadtxt("Ebptfeqrg.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))

#e,f = np.loadtxt("2f6jeqrg.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))

x,y1 = np.loadtxt("./DNA/MDNASASA1.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
y2 = np.loadtxt("./DNA/MDNASASA2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
y3 = np.loadtxt("./DNA/MDNASASA3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
y4 = np.loadtxt("./DNA/MDNASASA4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
y5 = np.loadtxt("./DNA/MDNASASA5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
y0 = np.concatenate((y1[-70000:],y2[-70000:],y3[-70000:],y4[-70000:],y5[-70000:]))

c,d1 = np.loadtxt("./DNA/EnoBDNASASA1.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
d2 = np.loadtxt("./DNA/EnoBDNASASA2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d3 = np.loadtxt("./DNA/EnoBDNASASA3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d4 = np.loadtxt("./DNA/EnoBDNASASA4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d5 = np.loadtxt("./DNA/EnoBDNASASA5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
d0 = np.concatenate((d1[-70000:],d2[-70000:],d3[-70000:],d4[-70000:],d5[-70000:]))

e,f1 = np.loadtxt("./DNA/EDNASASA1.dat", usecols=(0,1), unpack=True, comments=('#','@','&'))
f2 = np.loadtxt("./DNA/EDNASASA2.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f3 = np.loadtxt("./DNA/EDNASASA3.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f4 = np.loadtxt("./DNA/EDNASASA4.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f5 = np.loadtxt("./DNA/EDNASASA5.dat", usecols=(1), unpack=True, comments=('#','@','&'))
f0 = np.concatenate((f1[-70000:],f2[-70000:],f3[-70000:],f4[-70000:],f5[-70000:]))

plt.hist(y0, histtype='step', bins=300, label='NCP')
plt.hist(d0, histtype='step', bins=300, label='NCP + PHD')
plt.hist(f0, histtype='step', bins=300, label='NCP + BPTF')
plt.legend()
```
