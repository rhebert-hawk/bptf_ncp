```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y1=np.loadtxt('./fingerex_bonds/total1.dat', unpack=True, usecols=(0,1), comments=('#'))
y2=np.loadtxt('./fingerex_bonds/total2.dat', unpack=True, usecols=(1), comments=('#'))
y3=np.loadtxt('./fingerex_bonds/total3.dat', unpack=True, usecols=(1), comments=('#'))
y4=np.loadtxt('./fingerex_bonds/total4.dat', unpack=True, usecols=(1), comments=('#'))
y5=np.loadtxt('./fingerex_bonds/total5.dat', unpack=True, usecols=(1), comments=('#'))

y=np.loadtxt('fingerextotal.dat', unpack=True, usecols=(1), comments=('#'))

z1=np.loadtxt('./2f6jex_bonds/total1.dat', unpack=True, usecols=(1), comments=('#'))
z2=np.loadtxt('./2f6jex_bonds/total2.dat', unpack=True, usecols=(1), comments=('#'))
z3=np.loadtxt('./2f6jex_bonds/total3.dat', unpack=True, usecols=(1), comments=('#'))
z4=np.loadtxt('./2f6jex_bonds/total4.dat', unpack=True, usecols=(1), comments=('#'))
z5=np.loadtxt('./2f6jex_bonds/total5.dat', unpack=True, usecols=(1), comments=('#'))

z=np.loadtxt('2f6jextotal.dat', unpack=True, usecols=(1), comments=('#'))

a1=np.loadtxt('./EnoB_bonds/total1.dat', unpack=True, usecols=(1), comments=('#'))
a2=np.loadtxt('./EnoB_bonds/total2.dat', unpack=True, usecols=(1), comments=('#'))
a3=np.loadtxt('./EnoB_bonds/total3.dat', unpack=True, usecols=(1), comments=('#'))
a4=np.loadtxt('./EnoB_bonds/total4.dat', unpack=True, usecols=(1), comments=('#'))
a5=np.loadtxt('./EnoB_bonds/total5.dat', unpack=True, usecols=(1), comments=('#'))

a=np.loadtxt('EnoBtotal.dat', unpack=True, usecols=(1), comments=('#'))

b1=np.loadtxt('./E_bonds/total1.dat', unpack=True, usecols=(1), comments=('#'))
b2=np.loadtxt('./E_bonds/total2.dat', unpack=True, usecols=(1), comments=('#'))
b3=np.loadtxt('./E_bonds/total3.dat', unpack=True, usecols=(1), comments=('#'))
b4=np.loadtxt('./E_bonds/total4.dat', unpack=True, usecols=(1), comments=('#'))
b5=np.loadtxt('./E_bonds/total5.dat', unpack=True, usecols=(1), comments=('#'))

b=np.loadtxt('Etotal.dat', unpack=True, usecols=(1), comments=('#'))
```

```python
fig, axs = plt.subplots(4,1, sharex=True, sharey=True)

axs[0].plot(x/100,y1, alpha=0.5)
axs[0].plot(x/100,y2, alpha=0.5)
axs[0].plot(x/100,y3, alpha=0.5)
axs[0].plot(x/100,y4, alpha=0.5)
axs[0].plot(x/100,y5, alpha=0.5)

axs[1].plot(x/100,z1, alpha=0.5)
axs[1].plot(x/100,z2, alpha=0.5)
axs[1].plot(x/100,z3, alpha=0.5)
axs[1].plot(x/100,z4, alpha=0.5)
axs[1].plot(x/100,z5, alpha=0.5)

axs[2].plot(x/100,a1, alpha=0.5)
axs[2].plot(x/100,a2, alpha=0.5)
axs[2].plot(x/100,a3, alpha=0.5)
axs[2].plot(x/100,a4, alpha=0.5)
axs[2].plot(x/100,a5, alpha=0.5)

axs[3].plot(x/100,b1, alpha=0.5)
axs[3].plot(x/100,b2, alpha=0.5)
axs[3].plot(x/100,b3, alpha=0.5)
axs[3].plot(x/100,b4, alpha=0.5)
axs[3].plot(x/100,b5, alpha=0.5)
```

```python
plt.hist(y, alpha=0.5, histtype='step', label='PHD + Peptide')
plt.hist(z, alpha=0.5, histtype='step', label='BPTF + Peptide')
plt.hist(a, alpha=0.5, histtype='step', label='PHD + NCP')
plt.hist(b, alpha=0.5, histtype='step', label='BPTF + NCP')
plt.legend()
```
