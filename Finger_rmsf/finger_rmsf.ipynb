```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import matplotlib.legend_handler
font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'normal',
        'size': 12,
        }

x,y,yerr = np.loadtxt("Notnobrmsf.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
a,b,berr = np.loadtxt("E_finger.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
d,e,eerr= np.loadtxt("EnoBfingerrmsf.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
f,g,gerr= np.loadtxt("fingerexrmsf.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&'))
h,i,ierr= np.loadtxt("NTfinger.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&')) 
l,m,merr= np.loadtxt("2f6jfinger.dat", usecols=(0,1,2), unpack=True, comments=('#','@','&')) 

fig, axs = plt.subplots(2, figsize=(6.3, 3.5))
axs[0].set_title('PHD Finger RMSF', fontdict=font)
axs[0].plot(x, y, alpha=0.5, marker=".", lw=2, label='PHD', color='brown')
axs[0].plot(x, g, alpha=0.5, marker=".", lw=2, label='PHD + Pep', color='cyan')
axs[0].plot(x, e, alpha=0.5, marker=".", lw=2, label='PHD + NCP', color='pink')


axs[1].plot(a, i, alpha=0.5, marker=".", lw=2, label='BPTF', color='purple')
axs[1].plot(a, m, alpha=0.5, marker=".", lw=2, label='BPTF + Pep', color='orange')
axs[1].plot(a, b, alpha=0.5, marker=".", lw=2, label='BPTF + NCP', color='olive')
axs[0].fill_between(x, y-(yerr/2.2), y+(yerr/2.2), alpha=0.3, color='brown')
axs[0].fill_between(x, g-(gerr/2.2), g+(gerr/2.2), alpha=0.3, color='cyan')
axs[0].fill_between(x, e-(eerr/2.2), e+(eerr/2.2), alpha=0.3, color='pink')
axs[1].fill_between(a, b-(berr/2.2), b+(berr/2.2), alpha=0.3, color='olive')
axs[1].fill_between(a, i-(ierr/2.2), i+(ierr/2.2), alpha=0.3, color='purple')
axs[1].fill_between(a, m-(merr/2.2), m+(merr/2.2), alpha=0.3, color='orange')

axs[0].axvline(x=16.5, linestyle='dashed', color='black')
axs[0].axvline(x=48.5, linestyle='dashed', color='black')
axs[0].annotate('Binding Pocket', xy=(28,5), fontsize=8)

axs[1].axvline(x=16.5, linestyle='dashed', color='black')
axs[1].axvline(x=48.5, linestyle='dashed', color='black')
axs[1].annotate('Binding Pocket', xy=(28,5), fontsize=8)


leg = axs[0].legend(prop={'size': 8}, loc='upper left')
leg2 = axs[1].legend(prop={'size': 8}, loc='upper left')
axs[0].set_xlim([0,59])
axs[1].set_xlim([0,59])
axs[0].set_ylim([0,7])
axs[1].set_ylim([0,7])
axs[0].set_ylabel('RMSF ($\AA$)', fontdict=font)
axs[1].set_xlabel('Residue Number', fontdict=font)
axs[0].set_xticks(np.arange(0,59,5))
axs[0].set_xticklabels(np.arange(0,59,5), fontsize=10)
axs[1].set_xticks(np.arange(0,59,5))
axs[1].set_xticklabels(np.arange(0,59,5), fontsize=10)
axs[0].set_yticks(np.arange(0,8,1))
axs[1].set_yticks(np.arange(0,8,1))
axs[0].set_yticklabels(np.arange(0,8,1), fontsize=10)
axs[1].set_yticklabels(np.arange(0,8,1), fontsize=10)
plt.tight_layout()
plt.savefig('finger_rmsf.pdf')
```
